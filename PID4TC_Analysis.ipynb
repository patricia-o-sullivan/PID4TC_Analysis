```python
# List of Subject ID
subjects = range(1,21)
# List of Windows
windows = [2, 3, 4, 5, 6, 8, 10, 12, 15]

# This function loads the data for all the subjects and a given window (win)
def data_load(win = 2):
    df = pd.read_csv("/content/drive/MyDrive/PID4TC_Database/Sub" + str(subjects[0]) + "Features_L_" + str(win) + ".csv")
    for s in subjects[1:]:
        df = df.append(pd.read_csv("/content/drive/MyDrive/PID4TC_Database/Sub" + str(s) + "Features_L_" + str(win) + ".csv"))
    return df

```

#Evaluation of Random Forest Classifier

In the following code cell, we:
- load the data
- divide it in 10 folds that separate the subjects
- impute the missing values
- train and test the model on the different folds
- compute the SHAP analysis of the classifier
- aggregate the results for each window
- save the results as images

```python
# list of the final accuracy of the cross validation for each window
accuracies = []

# iterating over all the possible windows
for win in windows:
    print('#'*20)
    print('------ Window:',win)
    print('#'*20)

    # load the dataframe
    df = data_load(win)

    # establish CV scheme
    CV = KFold(n_splits=10, shuffle=False)

    # indices of rows for test and train
    ix_training, ix_test = [], []
    # loop through each fold and append the training & test indices to the empty lists above
    for fold in CV.split(df):
        ix_training.append(fold[0]), ix_test.append(fold[1])

    # the features are all the columns except for task, subject ID, and dominant hand
    X = df.drop(["Task", "SubID", "Dominant_Hand"] ,axis=1)
    selected_features_name = X.columns

    # the simple imputer is replacing missing values with the mean of the column
    imp = SimpleImputer(strategy="mean")
    X = pd.DataFrame(imp.fit_transform(X))

    # the target values
    y = df["Task"]
    # encoding the tasks with ids
    enc = preprocessing.LabelEncoder()
    y = pd.Series(enc.fit_transform(y))

    # a matrix collecting all the shap values
    SHAP_values_per_fold = [ [] for _ in range(len(enc.classes_))] #-#-#
    # confusion matrix for the given window
    win_cm = 0

    # loop through each outer fold to extract cm and SHAP values
    for i, (train_outer_ix, test_outer_ix) in enumerate(zip(ix_training, ix_test)): #-#-#
        print('\n------ Fold Number:',i)
        # train and test split
        X_train = X.iloc[train_outer_ix, :]
        X_test = X.iloc[test_outer_ix, :]
        y_train = y.iloc[train_outer_ix]
        y_test = y.iloc[test_outer_ix]

        # printing the unique subject ids and dominant hands for training and test
        print(df.iloc[train_outer_ix, -2].unique())
        print(df.iloc[test_outer_ix, -2].unique())
        print(df.iloc[train_outer_ix, -1].unique())
        print(df.iloc[test_outer_ix, -1].unique())

        # classifier definition
        model = RandomForestClassifier(random_state=10) # Random state for reproducibility (same results every time)

        # classifier training
        fit = model.fit(X_train, y_train)

        # classifier prediction
        y_pred = model.predict(X_test)

        print("The test accuracy is :" , model.score(X_test, y_test))

        # creation/addition to the confusion matrix
        if i == 0: # if it's the first iteration I create it
            win_cm = confusion_matrix(y_test, y_pred)
        else: # otherwise I add the new values
            win_cm += confusion_matrix(y_test, y_pred)

        # creation of the shap explainer
        explainer = shap.TreeExplainer(model)
        shap_values = explainer.shap_values(X_test)
        # saving the shap values in a matrix
        for j, c in enumerate(enc.classes_):
            for SHAPs in shap_values[j]:
                SHAP_values_per_fold[j].append(SHAPs) #-#-#
    # End KFold
    shap_val = []
    # aligning the shap matrices
    new_index = [ix for ix_test_fold in ix_test for ix in ix_test_fold]
    # printing the task specific shap analysis for all the different tasks
    for i, c in enumerate(enc.classes_):
        shap_val.append(np.array(SHAP_values_per_fold[i]))
        # create the shap plot
        shap.summary_plot(np.array(SHAP_values_per_fold[i]), X.reindex(new_index),plot_size=[12,10], feature_names=selected_features_name, class_names=enc.classes_, show= False, title= "Feature relevance for " + c + " window of " + str(win))
        # saving it as a picture
        plt.yticks(fontsize=20)
        plt.xticks(fontsize=20)
        plt.savefig("/content/drive/MyDrive/PID4TC_Images/feature_importance_" + c + "_" + str(win) + ".png", dpi=150)
        plt.close()
    # printing the feature importance shap analysis (bar chart)
    shap.summary_plot(shap_val, X.reindex(new_index), feature_names=selected_features_name, class_names=enc.classes_, plot_size=[12,7], show= False)
    plt.yticks(fontsize=18)
    plt.xticks(fontsize=18)
    plt.savefig("/content/drive/MyDrive/PID4TC_Images/feature_relevance_" + str(win) + ".png", dpi=150)
    plt.show()

    # computing the accuracy of the window confusion matrix
    accuracies.append(np.sum(np.diagonal(win_cm))/np.sum(win_cm))
    print("The overall Window Accuracy is: " , accuracies[-1])

    # printing and saving the confusion matrix
    disp = ConfusionMatrixDisplay(win_cm, display_labels= enc.classes_)
    plt.rcParams['font.size'] = 18
    disp.grid = False
    disp.plot(xticks_rotation = 'vertical',cmap=plt.cm.Blues)
    plt.grid(False)
    plt.savefig("/content/drive/MyDrive/PID4TC_Images/confusion_matrix_" + str(win) + ".png", dpi=150, bbox_inches= 'tight')
    plt.show()
```

```python
for i, win in enumerate(windows):
    print(win, "  ", round(accuracies[i],2))
```

```python
data = pd.DataFrame({"Window" : windows, "Accuracy" : accuracies })
data.plot.line(x = "Window", y = 'Accuracy', title = "Classification accuracy over window size", xlabel = "Window Size", ylabel = "Cross Validation Accuracy")
plt.gca().get_legend().remove()
plt.grid()
plt.rcParams['font.size'] = 18
plt.savefig("/content/drive/MyDrive/PID4TC_Images/accuracy_line_plot_allfeatures.png", dpi=150, bbox_inches= 'tight')
plt.show()
```
